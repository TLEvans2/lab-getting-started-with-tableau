```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')
display(data.shape)
data.head()
```

```python
gender = data.groupby(['Gender']).agg({'Customer':'count'})
```

```python
def discrete_plot(df):
    for column in df.columns:
        sns.countplot(x=column, data=df)
        plt.show()

test = discrete_plot(gender)
```

```python
# data.pivot_table(index= ['Gender','Employment Status'], values = ['','Profit','Discount']).reset_index()
df = data.pivot_table(index= ['Gender','EmploymentStatus'], values = ['Customer'], aggfunc = {'Customer':'count'}).reset_index()
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df'
# Create a bar plot
sns.barplot(x='EmploymentStatus', y='Customer', hue='Gender', data=df)
plt.xlabel('Employment Status')
plt.ylabel('Number of Customers')
plt.title('Number of Customers per Employment Status and Gender')
plt.show()
```

```python
def print_data_type_aliases(df):
    for column_name, data_type in df.dtypes.items():
        if data_type == 'object':
            print(f"{column_name}: Dimension")
        elif data_type == 'float64':
            print(f"{column_name}: Measure")
        else:
            print(f"{column_name}: Dimension")
```

```python
print_data_type_aliases(data)
```
